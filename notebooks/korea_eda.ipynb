```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pycaret as pcr
from utils.utils import get_project_root
from utils.df_z_score import z_score
```

```python
df = pd.read_csv(get_project_root() / 'data/Edited_Korea_Normalized.csv')
```

```python
df = df[['Measurement date', 'Station code', 'PM2.5']]
```

```python
pivot_df = df.pivot(index='Measurement date', columns='Station code', values='PM2.5')

# Переименование колонок для удобства (опционально)
pivot_df.columns.name = None
pivot_df.reset_index(inplace=True)
```

```python
pivot_df = pivot_df.drop('Measurement date', axis=1)
```

```python
df_zscored = z_score(pivot_df)
```
